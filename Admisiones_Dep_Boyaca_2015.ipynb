# Grafica Admitidos por programa en Boyacá

```python

# Configura el tamaño de la figura
plt.figure(figsize=(12, 8))

# Obtener la cuenta de valores en 'nombreprograma' y ordenar
facultades = Admitidos_2015_df_renombrado['nombreprograma'].value_counts()

# Seleccionar los 20 primeros resultados
top_20_facultades = facultades.head(20)

# Crear el gráfico de barras
ax = top_20_facultades.plot(kind='bar', color='Turquoise', edgecolor='purple')

# Umbral para mostrar anotaciones (puedes ajustar el umbral según tu necesidad)
umbral = 1000

# Añadir anotaciones en las barras que superen el umbral
for p in ax.patches:
    if p.get_height() > umbral:
        ax.annotate(f'{p.get_height()}', 
                    (p.get_x() + p.get_width() / 2., p.get_height()), 
                    ha='center', va='center', 
                    xytext=(0, 5), textcoords='offset points', 
                    rotation=90)

# Ajustar las etiquetas de los ejes y el título
plt.xticks(rotation=90, ha='right')
plt.title('Top 20 Admisiones por Programa')
plt.xlabel('Nombre del Programa')
plt.ylabel('Número de Admisiones')

# Ajustar el diseño para que no se corten los elementos
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

### Gráfica admisiones por Programa

- Podemos apreciar que en el departamento de boyaca contiene mayor número de admitidos en el programa de licenciatura en educación básica con énfasis en matemáticas
- Luego le sigue tecnología de obras civiles, tal ves por su posición demografica.
- Se muestra que a pesar de que Boyacá, es un  departamento agricultor no hay tanta población admitida para el programa de medicina veterinaria y zootecnía
- Se nota que la población esta muy interesada en el area de las matemáticas.

# Tendencia de estrato por modalidad y facultad

```python

# Definir las 20 principales facultades
top_20_facultades = Admitidos_2015_df_renombrado['Facultad'].value_counts().head(20).index.tolist()

# Filtrar el dataframe para mantener solo las 20 principales facultades
filtered_df = Admitidos_2015_df_renombrado[Admitidos_2015_df_renombrado['Facultad'].isin(top_20_facultades)]

# Agrupar por facultad y modalidad para obtener la suma de estrato
grouped_df = filtered_df.groupby(['Facultad', 'modalidad'])['estrato'].sum().unstack().fillna(0)

# Configurar el tamaño de la figura
plt.figure(figsize=(14, 10))

# Graficar el dataframe agrupado como un gráfico de líneas
grouped_df.plot(kind='line', colormap='tab20', marker='o', ax=plt.gca())

# Añadir título y etiquetas a los ejes
plt.title('Tendencia de Estrato por Modalidad y Facultad')
plt.xlabel('Facultad')
plt.ylabel('Suma de Estrato')
plt.xticks(rotation=90)  # Rotar etiquetas del eje x para mejor legibilidad

# Añadir leyenda
plt.legend(title='Modalidad', bbox_to_anchor=(1.05, 1), loc='upper left')

# Ajustar el diseño para que no se corten los elementos
plt.tight_layout()

# Mostrar la gráfica
plt.show()
```

+ Observamos que la tendencia de admitidos en Boyacá por cada facultad ha sido del estratosocioeconomico 2, y su modadlidad de estudio es presencial.

```python
# Filtrar los primeros 7 datos
data_subset = Admitidos_2015_df_renombrado.head(7)

# Definir el esquema de colores llamativos
palette = {'F': 'purple', 'M': 'blue'}

# Listas de gráficos para subgráficas de columnas
gráficos = {
    'Número de Estudiantes por Facultad': {
        'type': 'countplot',
        'x': 'Facultad',
        'hue': 'Genero',
        'title': 'Número de Estudiantes por Facultad'
    },
    'Número de Estudiantes por Jornada': {
        'type': 'countplot',
        'x': 'jornada',
        'hue': 'Genero',
        'title': 'Número de Estudiantes por Jornada'
    },
    'Número de Estudiantes por Modalidad': {
        'type': 'countplot',
        'x': 'modalidad',
        'hue': 'Genero',
        'title': 'Número de Estudiantes por Modalidad'
    },
    'Número de Estudiantes por Ciudad': {
        'type': 'countplot',
        'x': 'Ciudad',
        'hue': 'Genero',
        'title': 'Número de Estudiantes por Ciudad'
    }
}

# Determinar el número de subgráficas
n_gráficos = len(gráficos)
n_cols = 2  # Puedes ajustar esto según la cantidad de subgráficas y el diseño deseado
n_rows = (n_gráficos + n_cols - 1) // n_cols  # Cálculo para el número de filas

# Crear la figura y los ejes
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5 * n_rows), sharey=False)
axes = axes.flatten()  # Aplanar el array de ejes para iterar fácilmente

# Crear una subgráfica para cada gráfico
for ax, (titulo, params) in zip(axes, gráficos.items()):
    if params['type'] == 'countplot':
        # Crear gráficos de columnas
        sns.countplot(data=data_subset, 
                      x=params['x'], 
                      hue=params['hue'], 
                      palette=palette, 
                      ax=ax)
        
    ax.set_title(params['title'], fontsize=14, fontweight='bold')
    ax.set_xlabel(params['x'], fontsize=12)
    ax.set_ylabel('Número de Estudiantes', fontsize=12)
    ax.legend(title='Genero', title_fontsize='13', fontsize='11')  # Ajustar leyenda

    # Mejorar la legibilidad de los ticks del eje x
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')

# Eliminar los ejes no utilizados si hay menos gráficos que subplots
for ax in axes[len(gráficos):]:
    ax.remove()

# Ajustar el diseño para una visualización clara
plt.tight_layout()
plt.show()
```
